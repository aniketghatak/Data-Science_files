```python
import pandas as pd
```

```python
df=pd.DataFrame({"Size":["s","m","l","xl","s","m","l","m","l","xl","m"]})
df
```

```python
ordinal_data=[["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=ordinal_data)
oe.fit(df[["Size"]])
```

```python
oe.transform(df[["Size"]])
```

```python
df["size_new"]=oe.transform(df[["Size"]])
```

```python
df
```

#### Map 

```python
ordinal_data2={"s":0,"m":1,"l":2,"xl":3}
```

```python
df["Size_map_new"]=df["Size"].map(ordinal_data2)
```
