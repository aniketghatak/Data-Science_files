```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv(r"D:\Data Analysis Projects\DS_Playlist\datasets\loan_data_set.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
sns.distplot(dataset["CoapplicantIncome"])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms=MinMaxScaler()
ms.fit(dataset[["CoapplicantIncome"]])
```

```python
ms.transform(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_min"]=ms.transform(dataset[["CoapplicantIncome"]])
```

```python
dataset.head(10)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["CoapplicantIncome"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["CoapplicantIncome_min"])

plt.show()
```
